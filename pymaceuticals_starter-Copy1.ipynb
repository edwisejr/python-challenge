```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"


# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
ct_data = pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset
full_data = pd.merge(mouse_data,ct_data,how='outer', on ='Mouse ID' )

#full_data = full_data .loc[full_data ['Drug'].isin(['Capomulin', 'Infubinol', 'Ketapril','Placebo'] )]
# Display the data table for preview
full_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_time = pd.DataFrame (full_data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean().reset_index())
# Convert to DataFrame
# Preview DataFrame
drug_time.head(25)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_time_sem  = pd.DataFrame (full_data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem().reset_index())
# Convert to DataFrame
# Preview DataFrame
tumor_time_sem.head()
tumor_time_sem
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_pivot = drug_time.pivot( index = 'Timepoint', columns='Drug',values='Tumor Volume (mm3)')
# Preview that Reformatting worked

mean_pivot
```

```python
sem_pivot =  tumor_time_sem.pivot( index = 'Timepoint', columns='Drug',values='Tumor Volume (mm3)')
sem_pivot 
```

```python
x_axis =  mean_pivot.index.values
formats= ['-ro','-g+','-b*','-k.','-yx','-rs','-gv','-mh','-ys','-kd']
for i in range(len(mean_pivot)):
    plt.errorbar( x_axis , mean_pivot.iloc[:,i], sem_pivot.iloc[:,i],fmt=formats[i])
i += 1
    
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid(True)
plt.legend(bbox_to_anchor=(0, .8),
           bbox_transform=plt.gcf().transFigure)
# Save the Figure
plt.savefig('tumor_size_plot.png', transparent=True, bbox_inches='tight', pad_inches=0)
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint
mean_met= pd.DataFrame (full_data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean().reset_index())
# Convert to DataFrame
# Preview DataFrame
mean_met
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
tumor_met_sem  = pd.DataFrame (full_data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem().reset_index())
# Convert to DataFrame
# Preview DataFrame
tumor_met_sem
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_met_pivot = mean_met.pivot( index = 'Timepoint', columns='Drug',values='Metastatic Sites')
mean_met_pivot
# Preview that Reformatting worked
```

```python
sem_met_pivot = tumor_met_sem.pivot( index = 'Timepoint', columns='Drug',values='Metastatic Sites')
sem_met_pivot
```

```python
# Generate the Plot (with Error Bars)
x_axis = mean_met_pivot.index.values
formats= ['-ro','-g+','-b*','-k.','-yx','-rs','-gv','-mh','-ys','-kd']
for i in range(len(mean_met_pivot)):
    plt.errorbar( x_axis , mean_met_pivot.iloc[:,i], sem_met_pivot.iloc[:,i],fmt=formats[i])
i += 1
    
plt.title("Metastatic Spreading During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Met. Sites")
plt.grid(True)
plt.legend(bbox_to_anchor=(0, .8),
           bbox_transform=plt.gcf().transFigure)
# Save the Figure
plt.savefig('met_plot.png', transparent=True, bbox_inches='tight', pad_inches=0)
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = pd.DataFrame (full_data.groupby(['Drug', 'Timepoint'])['Mouse ID'].nunique().reset_index())
# Convert to DataFrame
# Preview DataFrame
mouse_count 
```

```python
# Minor Data Munging to Re-Format the Data Frames
count_pivot = mouse_count.pivot( index = 'Timepoint', columns='Drug',values='Mouse ID')
# Preview the Data Frame
count_pivot
```

```python
percent_change = count_pivot.div(25, axis='index')
percent_change  = percent_change*100
percent_change
```

```python
# Generate the Plot (Accounting for percentages)
x_axis = percent_change.index.values
colors = ['r','g','b','k','y','r','g','m','y','k']
marker= ['o','+','*','.','x','s','v','h','s','d']
for i in range(len(percent_change)):
    plt.plot( x_axis , percent_change.iloc[:,i],color=colors[i], marker=marker[i])
i += 1

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.grid(True)
plt.legend(bbox_to_anchor=(0, .8),
           bbox_transform=plt.gcf().transFigure)

# Save the Figure
plt.savefig('survival_plot.png', transparent=True, bbox_inches='tight', pad_inches=0)
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
vol_change = mean_pivot.loc[[0,45]].pct_change()*100
# Display the data to confirm
vol_change = vol_change.loc[45]
vol_change
```

```python

colors = []
N = len(vol_change)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

i =0
for i in range(N):
    if vol_change[i] <= 0:
        colors.append('g')
    else:
        colors.append('r')
i += 1

plt.bar( ind,vol_change, color = colors)

plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Change Tumor Volume")
plt.grid(True)

plt.show()
```
